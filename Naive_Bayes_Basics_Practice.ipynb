**Problem :
outlook - Sunny ; temp - Hot ; humidty - High ; wind - Weak
Predict - Play/No Play

**Solution:

P(yes|sunny,hot,high,weak) = P(sunny,hot,high,weak|yes)*P(yes)/P(sunny,hot,high,weak) = p1

P(no|sunny,hot,high,weak) = P(sunny,hot,high,weak|no)*P(no)/P(sunny,hot,high,weak) = p2

As, P(sunny,hot,high,weak) term is common in the denominator of both YES and NO hence it is ignored from our computation

```python
# First we find out the P(yes) & P(no)
```
